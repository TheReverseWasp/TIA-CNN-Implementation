# Data agmentation in Animal Image Dataset
The present project adapts Animal Image dataset from https://www.kaggle.com/ashishsaxena2209/animal-image-datasetdog-cat-and-panda to the dimensions of LFW Dataset on https://www.kaggle.com/jessicali9530/lfw-dataset

## 0. Common Imports

```python
import pandas as pd
import os
import numpy as np
from PIL import Image
from matplotlib.pyplot import imshow
```

## 1. Find the dimension of LFW

```python
data_folder = "./Data/Data/"
face_folder = "lfw-deepfunneled/lfw-deepfunneled/"
name_ds = pd.read_csv(data_folder + "lfw/lfw_allnames.csv")
name_list = name_ds.values.tolist()
name = name_list[0][0]
name
```

```python
image_test = Image.open(data_folder + "lfw/" + face_folder+ name + "/" + name +"_0001.jpg")
sz = image_test.size
print(sz)
imshow(np.asarray(image_test))
```

## 2. Resize all AID images to this format

```python
#helper function
def converter_string(num):
    var = str(num)
    while len(var) < 5:
        var = "0" + var
    var = "_" + var
    return var
```

```python
animal_folder = "aid/animals/"
aug_folder = "aid_aug/animals/"
animals = ["cats", "dogs", "panda"]
for i in animals:
    print("Processing: ", i)
    for j in range(1, 1001):
        filename = data_folder + animal_folder + i + "/" + i + converter_string(j) +".jpg"
        file_destiny = data_folder + aug_folder + i + "/" + i + converter_string(j) +".jpg"
        im = Image.open(filename)
        save_im = im.resize(sz)
        save_im.save(file_destiny)
print("finished")
```
